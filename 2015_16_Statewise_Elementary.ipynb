```python
gb= df.groupby("STATNAME").agg({"DISTRICTS":'mean',"VILLAGES":'mean',"POPULATION_0_6":'mean'})
print(gb)
```

```python
plt.figure(figsize=(20,30))
sns.heatmap(gb,annot=True)
plt.show()
```
